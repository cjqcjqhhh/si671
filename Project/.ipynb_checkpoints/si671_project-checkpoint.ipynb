```python
import numpy as np
import pandas as pd
import json
import seaborn as sns
from matplotlib import pyplot as plt
import warnings
warnings.filterwarnings('ignore')
warnings.simplefilter('ignore')
```

## Load Data

```python
# load LOL game data
games = pd.read_csv("./data/games.csv")
games.head()
```

```python
# load champion data
with open('data/champion_info_2.json','r') as f:
    champ_data = json.loads(f.read())

champ_table = pd.DataFrame.from_dict(champ_data["data"], orient="index")
champ_table = champ_table.set_index('id').sort_index()
if -1 in champ_table.index:
    champ_table.drop(-1)
champ_dict = {index: row["name"] for index, row in champ_table.iterrows()}
champ_table.head()
```

```python
# define some additional useful functions here
def get_champ_name(i, champ_dict = champ_dict):
    return champ_dict[i]
```

## Frequent champion combination

### Task 0: Champion statistical analysis

```python
from collections import Counter
from collections import defaultdict

pick_key = [
    "t1_champ1id", "t1_champ2id", "t1_champ3id", "t1_champ4id", "t1_champ5id", 
    "t2_champ1id", "t2_champ2id", "t2_champ3id", "t2_champ4id", "t2_champ5id",
]
ban_key = [
    't1_ban1', 't1_ban2', 't1_ban3', 't1_ban4', 't1_ban5',
    't2_ban1', 't2_ban2', 't2_ban3', 't2_ban4', 't2_ban5',
]

# count pick rate
pick_list = games[pick_key].values.reshape(-1)
champ_pick = Counter(pick_list)
champ_pick_rate = {k: v / len(pick_list) for k, v in champ_pick.items()}

# count ban rate
ban_list = games[ban_key].values.reshape(-1)
champ_ban = Counter(ban_list)
champ_ban_rate = {k: v / len(pick_list) for k, v in champ_ban.items()}

# generate statistical dataframe
champ_statistic_df = pd.DataFrame.from_dict([champ_dict, champ_pick_rate, champ_ban_rate]).transpose()
champ_statistic_df.columns = ["Name", "Pick Rate", "Ban Rate"]
if -1 in champ_statistic_df.index:
    champ_statistic_df.drop(-1)

# make plot
```

```python
fig, axes = plt.subplots(1, 2, figsize=(16, 4))
# fig.suptitle('LoL Champion Statistics (Top 5 pick and ban)')
sns.barplot(
    ax=axes[0],
    data=champ_statistic_df.sort_values("Pick Rate", ascending=False).head(5), 
    x="Name", 
    y="Pick Rate"
)
sns.barplot(
    ax=axes[1],
    data=champ_statistic_df.sort_values("Ban Rate", ascending=False).head(5), 
    x="Name", 
    y="Ban Rate"
)
```

### Task 1: Explore on the frequent champion combinations on one side (in the same team)

```python
# data preprocessing
team1_champ_key = ["t1_champ1id", "t1_champ2id", "t1_champ3id", "t1_champ4id", "t1_champ5id"]
team2_champ_key = ["t2_champ1id", "t2_champ2id", "t2_champ3id", "t2_champ4id", "t2_champ5id"]

# convert to itemsets
champ_itemsets = \
    [set(row) for index, row in games[team1_champ_key].iterrows()] + \
    [set(row) for index, row in games[team2_champ_key].iterrows()]
champ_itemsets = champ_itemsets[0:20000]
# len(team_champ_itemsets)

# one-hot encoding
from sklearn.preprocessing import MultiLabelBinarizer
mlb = MultiLabelBinarizer()
mlb.fit(champ_itemsets)
champ_binary = pd.DataFrame(mlb.transform(champ_itemsets), columns=mlb.classes_)
champ_binary.head()
```

```python
# perform Apriori algorithm
from mlxtend.frequent_patterns import apriori

def team_champ_frequent_itemsets(k, min_support, dataset = champ_binary):
    frequent_itemsets = apriori(dataset, min_support = min_support, use_colnames=True)
    frequent_itemsets['length'] = frequent_itemsets['itemsets'].apply(lambda x: len(x))
    return frequent_itemsets[ (frequent_itemsets['length'] == k) &
                   (frequent_itemsets['support'] >= min_support) ].drop(columns=['length'])
```

```python
result = team_champ_frequent_itemsets(3, 0.0015)
result = result.sort_values(by = ['support'], ascending=False)
result["name"] = [
    [get_champ_name(i) for i in itemset] for itemset in result["itemsets"]
]
result[:10]
```

### Task 2: Explore on the frequent champion combinations on counter side (in the opposite teams)

```python
# data preprocessing
team1_champ_key = ["t1_champ1id", "t1_champ2id", "t1_champ3id", "t1_champ4id", "t1_champ5id"]
team2_champ_key = ["t2_champ1id", "t2_champ2id", "t2_champ3id", "t2_champ4id", "t2_champ5id"]

# convert to itemsets
team_champ_df = games[team1_champ_key+team2_champ_key]
team_champ_itemsets = []
for index, row in team_champ_df.iterrows():
    new_itemset = []
    for i, c in enumerate(row):
        if i < 5:
            new_itemset.append((c, 1))
        else:
            new_itemset.append((c, 2))
    team_champ_itemsets.append(set(new_itemset))
team_champ_itemsets = team_champ_itemsets[0:5000]
    
# one-hot encoding
from sklearn.preprocessing import MultiLabelBinarizer
mlb = MultiLabelBinarizer()
mlb.fit(team_champ_itemsets)
team_champ_binary = pd.DataFrame(mlb.transform(team_champ_itemsets), columns=mlb.classes_)
team_champ_binary.head()
```

```python
result = team_champ_frequent_itemsets(2, 0.01, dataset = team_champ_binary)
result = result.sort_values(by = ['support'], ascending=False)

# drop result of same team
drop_indice = []
for index, row in result.iterrows():
    itemset = list(row["itemsets"])
    if itemset[0][1] == itemset[1][1]: # in the same team
        drop_indice.append(index)

dropped_result = result.drop(index=drop_indice)
dropped_result["name"] = [
    [get_champ_name(i[0]) for i in itemset] for itemset in dropped_result["itemsets"]
]
dropped_result[:10]
```

## Winning Prediction

### Task: Predict the winning rate of a match based on team champion combination

```python
# data preprocessing
from sklearn.model_selection import train_test_split

features = [
    "t1_champ1id", "t1_champ2id", "t1_champ3id", "t1_champ4id", "t1_champ5id",
    "t2_champ1id", "t2_champ2id", "t2_champ3id", "t2_champ4id", "t2_champ5id",
    't1_ban1', 't1_ban2', 't1_ban3', 't1_ban4', 't1_ban5',
    't2_ban1', 't2_ban2', 't2_ban3', 't2_ban4', 't2_ban5',
]
X = games[features]
y = games["winner"]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

### Baseline

```python
# baseline 1
from sklearn.dummy import DummyClassifier
baseline1 = DummyClassifier(strategy="uniform").fit(X_train, y_train)
# baseline 2
from sklearn.linear_model import LogisticRegression
baseline2 = LogisticRegression().fit(X_train, y_train)

baseline1.score(X_test, y_test), baseline2.score(X_test, y_test)
```

### Feature selection

```python
# classify champions according to game duration (for only winning team)
from collections import defaultdict

champ_duration_dict = defaultdict(list)
for index, row in games.iterrows():
    candidate_key = team1_champ_key if row["winner"] == 1 else team2_champ_key
    for champ in row[candidate_key]:
        champ_duration_dict[champ].append(row["gameDuration"])   
avg_champ_duration_dict = {k: np.mean(v) for k, v in champ_duration_dict.items()}

sorted_avg_champ_duration_dict = sorted(avg_champ_duration_dict.items(), key=lambda kv: kv[1])
n = len(sorted_avg_champ_duration_dict)
early_champ = [x[0] for x in sorted_avg_champ_duration_dict[0:n//3]]
mid_champ = [x[0] for x in sorted_avg_champ_duration_dict[n//3:2*n//3]]
late_champ = [x[0] for x in sorted_avg_champ_duration_dict[2*n//3:]]
```

```python
# perform feature selection
t1_earlyChampCount = []
t2_earlyChampCount = []
t1_midChampCount = []
t2_midChampCount = []
t1_lateChampCount = []
t2_lateChampCount = []
for index, row in games.iterrows():
    t1_early = t2_early = t1_mid = t2_mid = t1_late = t2_late = 0
    for champ in row[team1_champ_key]:
        if champ in early_champ:
            t1_early += 1
        elif champ in mid_champ:
            t1_mid += 1
        elif champ in late_champ:
            t1_late += 1
        else:
            print("ERROR")
    for champ in row[team2_champ_key]:
        if champ in early_champ:
            t2_early += 1
        elif champ in mid_champ:
            t2_mid += 1
        elif champ in late_champ:
            t2_late += 1
        else:
            print("ERROR")
    t1_earlyChampCount.append(t1_early)
    t2_earlyChampCount.append(t2_early)
    t1_midChampCount.append(t1_mid)
    t2_midChampCount.append(t2_mid)
    t1_lateChampCount.append(t1_late)
    t2_lateChampCount.append(t2_late)

games["t1_earlyChampCount"] = t1_earlyChampCount
games["t2_earlyChampCount"] = t2_earlyChampCount
games["t1_midChampCount"] = t1_midChampCount
games["t2_midChampCount"] = t2_midChampCount
games["t1_lateChampCount"] = t1_lateChampCount
games["t2_lateChampCount"] = t2_lateChampCount
```

```python
# add additional features
additional_features = [
    'gameDuration', 'firstBlood', 'firstTower', 'firstDragon', 'firstRiftHerald',
    "t1_earlyChampCount", "t2_earlyChampCount", "t1_midChampCount", "t2_midChampCount", "t1_lateChampCount", "t2_lateChampCount", 
]
X = games[features + additional_features]
y = games["winner"]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

### Train the models

```python
from sklearn.model_selection import GridSearchCV
# model candidates dict
models = {}

# Random Forest
from sklearn.ensemble import RandomForestClassifier
param_grid = {
    "max_depth": [2, 3, 5]
}
rf = RandomForestClassifier(random_state=0)
clf = GridSearchCV(rf, param_grid).fit(X_train, y_train)
models["RandomForest"] = [clf.score(X_test, y_test), clf.best_params_]
```

```python
# Stochastic Gradient Boosting
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import SGDClassifier
param_grid = {
    "clf__penalty": ['l2', 'l1'],
    "clf__max_iter": [500, 1000, 2000], 
    "clf__alpha": [1e-4, 1e-3, 1e-2]
}
pipe = Pipeline([
    ('scaler', StandardScaler()),
    ('clf', SGDClassifier())
])
clf = GridSearchCV(pipe, param_grid).fit(X_train, y_train)
models["SGD"] = [clf.score(X_test, y_test), clf.best_params_]
```

```python
# Support Vector Machine
from sklearn.svm import LinearSVC
from sklearn.preprocessing import MinMaxScaler
param_grid = {
    "clf__C": [0.1, 1, 10, 100],
}
pipe = Pipeline([
    ('scaler', MinMaxScaler()),
    ('clf', LinearSVC())
])
clf = GridSearchCV(pipe, param_grid).fit(X_train, y_train)
models["LinearSVC"] = [clf.score(X_test, y_test), clf.best_params_]
```
